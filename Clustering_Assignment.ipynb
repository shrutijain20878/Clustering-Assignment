# Clustering Assignment- Theoretical

1.  **What is unsupervised learning in the context of machine learning?**

    Unsupervised learning is a machine learning approach where the model learns patterns from unlabeled data. It aims to find hidden structures, such as groups (clustering) or important features (dimensionality reduction), without predefined outputs.

2. **How does K-Means clustering algorithm work?**

    K-Means is an unsupervised learning algorithm that groups data into K clusters based on similarity.

    It works by:

    1. Randomly selecting K centroids.

    2. Assigning each data point to the nearest centroid (forming clusters).

    3. Updating each centroid to be the mean of points in its cluster.

    4. Repeating steps 2–3 until centroids stop changing or convergence is reached.

    The goal is to minimize the within-cluster variance (distance between points and their centroids).

3. **Explain the concept of a dendrogram in hierarchical clustering?**

   A dendrogram is a tree-like diagram used in hierarchical clustering to show how data points are merged or split at different distance levels.

   Each leaf represents a data point, and branches show how clusters combine as the distance threshold increases. By cutting the dendrogram at a chosen height, you can decide the number of clusters. It visually illustrates the hierarchy and relationships between clusters.

4. **What is the main difference between K-Means and Hierarchical Clustering?**

    The main difference is that K-Means requires you to predefine the number of clusters (K), while Hierarchical Clustering builds a tree-like structure (dendrogram) and lets you choose clusters later by cutting the tree.

    K-Means is faster and better for large datasets, whereas Hierarchical Clustering is more interpretable but computationally heavier for big data.

5. **What are the advantages of DBSCAN over K-Means**?

    DBSCAN has several advantages over K-Means:

    It does not require specifying the number of clusters beforehand.

    It can find clusters of arbitrary shapes, not just spherical ones.

    It effectively handles noise and outliers by labeling them as separate points.

    It works well when clusters have different sizes and densities, where K-Means may fail.

6. **When would you use Silhouette Score in clustering?**

    We use the Silhouette Score to evaluate the quality of clustering.

    It measures how similar each data point is to its own cluster compared to other clusters. The score ranges from -1 to 1 — values close to 1 indicate well-separated, compact clusters, while values near 0 or negative suggest overlapping or incorrect clustering.

    It helps in choosing the optimal number of clusters and comparing different clustering algorithms.

7. **What are the limitations of Hierarchical Clustering?**

    Hierarchical Clustering has a few key limitations:

    It is computationally expensive and slow for large datasets.

    Once a merge or split is made, it cannot be undone (no reassignments).

    It is sensitive to noise and outliers, which can distort the hierarchy.

    Choosing the right linkage method and number of clusters can be difficult.

8. **Why is feature scaling important in clustering algorithms like K-Means?**

    Feature scaling is important in clustering algorithms like K-Means because they rely on distance calculations (e.g., Euclidean distance) to form clusters.

    If features are on different scales, variables with larger ranges can dominate the distance measure, leading to biased or incorrect clusters. Scaling ensures all features contribute equally, improving clustering accuracy and fairness.

9. **How does DBSCAN identify noise points?**

    DBSCAN identifies noise points (outliers) based on density.

    A point is labeled as noise if it is not within the neighborhood of any core point, meaning it has fewer than min_samples points within a distance of eps. These points do not belong to any cluster and are treated as anomalies, allowing DBSCAN to separate clusters from sparse regions.

10. **Define inertia in the context of K-Means?**

    In K-Means, inertia is the sum of squared distances between each data point and the centroid of its assigned cluster.

    It measures how tightly the data points are grouped within clusters — lower inertia indicates more compact, well-defined clusters. It is often used to evaluate clustering quality or to help choose the optimal number of clusters using the elbow method.

11. **What is the elbow method in K-Means clustering?**
  
    The Elbow Method is a technique to choose the optimal number of clusters (K) in K-Means.

    It works by plotting the inertia (sum of squared distances to cluster centroids) against different values of K. Initially, inertia decreases sharply as K increases, but after a certain point, the rate of decrease slows down, forming an “elbow” in the plot. The K at this elbow point is considered optimal, balancing cluster compactness and simplicity.

12. **Describe the concept of "density" in DBSCAN?**

    In DBSCAN, density refers to the number of points within a given radius (eps) around a data point.

    A point is a core point if its neighborhood contains at least min_samples points.

    A dense region is formed by connected core points, creating a cluster.

    Points in sparse regions with fewer neighbors are labeled as noise.

    Density is the key idea that allows DBSCAN to detect clusters of arbitrary shapes and separate outliers.

13. **Can hierarchical clustering be used on categorical data?**
  
    Yes, hierarchical clustering can be used on categorical data, but you cannot use standard distance metrics like Euclidean.

    Instead, you use similarity or dissimilarity measures suitable for categorical data, such as:

    Hamming distance (counts mismatches between categories)

    Gower distance (handles mixed numerical and categorical features)

    Once the distance matrix is computed, hierarchical clustering (agglomerative or divisive) can build the dendrogram as usual.

14. **What does a negative Silhouette Score indicate?**
   
    A negative Silhouette Score indicates that a data point is closer to points in another cluster than to points in its own cluster.

    In other words, the point may be misclassified or poorly clustered, and the clusters are likely overlapping or not well-separated. Negative scores suggest the clustering solution is not appropriate for the data.

15. **Explain the term "linkage criteria" in hierarchical clustering?**
  
    In hierarchical clustering, linkage criteria determine how the distance between clusters is calculated when merging them.

    Common methods include:

    Single linkage: Distance between the closest points of two clusters.

    Complete linkage: Distance between the farthest points of two clusters.

    Average linkage: Average distance between all pairs of points in the clusters.

    Ward’s linkage: Minimizes the increase in total within-cluster variance after merging.

    The choice of linkage affects the shape and structure of the resulting dendrogram and clusters.

16. **Why might K-Means clustering perform poorly on data with varying cluster** **sizes or densities?**

    K-Means can perform poorly on data with varying cluster sizes or densities because it assumes clusters are spherical and of similar size. When clusters differ, K-Means may split large clusters or merge small, dense ones incorrectly, leading to inaccurate clustering. Density-based algorithms like DBSCAN handle such cases better.


17. **What are the core parameters in DBSCAN, and how do they influence clustering**

    The core parameters in DBSCAN are eps and min_samples.

    eps defines the radius around a point to search for neighbors. A small eps may result in many points being labeled as noise, while a large eps can merge distinct clusters.

    min_samples is the minimum number of points required to form a dense region (core point). Higher values make clusters stricter and reduce noise points, while lower values create more clusters but may include sparse regions.

    Together, these parameters control cluster density, shape, and the identification of noise.

18.  **How does K-Means++ improve upon standard K-Means initialization?**

      K-Means++ improves standard K-Means by using a smarter method to initialize centroids.
      
      Instead of choosing centroids randomly, K-Means++ selects the first centroid randomly and then picks subsequent centroids probabilistically based on their distance from existing centroids. This spreads centroids farther apart, reducing the chances of poor initial placement, leading to faster convergence and often better clustering results with lower inertia.

19. **What is agglomerative clustering**?

    Agglomerative clustering is a type of hierarchical clustering that builds clusters in a bottom-up approach.

    It starts with each data point as its own cluster and iteratively merges the closest clusters based on a chosen distance metric and linkage criterion. This process continues until all points are merged into a single cluster or a stopping condition is met. The results can be visualized using a dendrogram, which shows the hierarchy of cluster merges.

20. **What makes Silhouette Score a better metric than just inertia for model evaluation?**

    The Silhouette Score is often a better metric than inertia because it evaluates clustering quality from both cohesion and separation, not just compactness.

    While inertia measures the sum of squared distances within clusters, it only captures how tight clusters are and tends to decrease as the number of clusters increases, even if clusters overlap. In contrast, the Silhouette Score considers how close points are to their own cluster versus other clusters, giving a value between -1 and 1 that reflects both cluster separation and compactness, making it more informative for assessing clustering performance.

```python
# 29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart

import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

X, _ = make_blobs(n_samples=300, centers=4, cluster_std=1.0, random_state=42)

k_values = range(2, 6)
sil_scores = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    sil_scores.append(score)

plt.figure(figsize=(8,5))
plt.bar(k_values, sil_scores, color='skyblue')
plt.xticks(k_values)
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for Different k in K-Means')
plt.show()
```

```python
# 31 . Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=2.0, random_state=42)

kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)
centroids = kmeans.cluster_centers_

# Create a mesh grid for decision boundary visualization
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.05),
                     np.arange(y_min, y_max, 0.05))

# Predict cluster for each point in the grid
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot decision boundaries
plt.figure(figsize=(8,6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='tab10')
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='tab10', edgecolor='k')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title('K-Means Clustering with Overlapping Clusters')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
# 34.  Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a line plot

import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Load Breast Cancer dataset
cancer = load_breast_cancer()
X = cancer.data

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Compute inertia for K = 2 to 6
k_values = range(2, 7)
inertia_values = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia_values.append(kmeans.inertia_)

# Plot inertia vs K
plt.figure(figsize=(8,5))
plt.plot(k_values, inertia_values, marker='o', linestyle='--', color='b')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('K-Means Inertia for Breast Cancer Dataset')
plt.xticks(k_values)
plt.show()
```

```python
# 42.  Load the Breast Cancer dataset, reduce dimensionality using PCA, and apply Agglomerative Clustering. Visualize in 2D

import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering

# Load Breast Cancer dataset
cancer = load_breast_cancer()
X = cancer.data

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Reduce dimensionality to 2D using PCA
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)

# Apply Agglomerative Clustering
agg_cluster = AgglomerativeClustering(n_clusters=2, linkage='ward')
labels = agg_cluster.fit_predict(X_pca)

# Plot clusters in 2D
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis', s=50, edgecolor='k')
plt.title('Agglomerative Clustering on Breast Cancer Dataset (PCA-reduced)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
# 43. Generate noisy circular data using make_circles and visualize clustering results from KMeans and DBSCAN side-by-side
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import KMeans, DBSCAN

# Generate noisy circular data
X, _ = make_circles(n_samples=300, noise=0.05, factor=0.5, random_state=42)

# Apply K-Means
kmeans = KMeans(n_clusters=2, random_state=42)
labels_kmeans = kmeans.fit_predict(X)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.1, min_samples=5)
labels_dbscan = dbscan.fit_predict(X)

# Plot side-by-side
fig, axes = plt.subplots(1, 2, figsize=(14,6))

# K-Means plot
axes[0].scatter(X[:, 0], X[:, 1], c=labels_kmeans, cmap='viridis', s=50, edgecolor='k')
axes[0].scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='X', s=200, label='Centroids')
axes[0].set_title('K-Means Clustering')
axes[0].set_xlabel('Feature 1')
axes[0].set_ylabel('Feature 2')
axes[0].legend()

# DBSCAN plot
axes[1].scatter(X[labels_dbscan != -1, 0], X[labels_dbscan != -1, 1], c=labels_dbscan[labels_dbscan != -1], cmap='viridis', s=50, edgecolor='k', label='Clustered points')
axes[1].scatter(X[labels_dbscan == -1, 0], X[labels_dbscan == -1, 1], c='red', marker='x', s=50, label='Outliers')
axes[1].set_title('DBSCAN Clustering')
axes[1].set_xlabel('Feature 1')
axes[1].set_ylabel('Feature 2')
axes[1].legend()

plt.show()
```

```python
# 44. Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score

# Load Iris dataset
iris = load_iris()
X = iris.data

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X_scaled)

# Compute silhouette scores for each sample
sample_silhouette_values = silhouette_samples(X_scaled, labels)

# Plot silhouette scores
plt.figure(figsize=(8,6))
y_lower = 10
for i in range(3):
    ith_cluster_silhouette_values = sample_silhouette_values[labels == i]
    ith_cluster_silhouette_values.sort()
    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i

    plt.fill_betweenx(np.arange(y_lower, y_upper),
                      0, ith_cluster_silhouette_values,
                      alpha=0.7)
    plt.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
    y_lower = y_upper + 10  # 10 for spacing between clusters

plt.xlabel("Silhouette Coefficient")
plt.ylabel("Cluster label")
plt.title("Silhouette Coefficient for each sample - Iris Dataset")
plt.axvline(x=silhouette_score(X_scaled, labels), color="red", linestyle="--", label="Average Silhouette Score")
plt.legend()
plt.show()
```

```python
# 46. Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4 features)

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Load Wine dataset
wine = load_wine()
X = wine.data
y = wine.target  # optional, original labels
feature_names = wine.feature_names

# Create DataFrame with first 4 features
df = pd.DataFrame(X[:, :4], columns=feature_names[:4])

# Scale features
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

# Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(df_scaled)

# Add cluster labels to DataFrame
df_scaled['Cluster'] = labels.astype(str)  # as string for hue in pairplot

# Pairplot
sns.pairplot(df_scaled, hue='Cluster', palette='tab10', diag_kind='kde')
plt.suptitle('K-Means Clustering on Wine Dataset (First 4 Features)', y=1.02)
plt.show()
```

```python
# 47. Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the count

from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import numpy as np

# Generate noisy blobs
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=1.5, random_state=42)

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply DBSCAN
dbscan = DBSCAN(eps=1.2, min_samples=5)
labels = dbscan.fit_predict(X_scaled)

# Count clusters (excluding noise)
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
# Count noise points
n_noise = np.sum(labels == -1)

print("Number of clusters (excluding noise):", n_clusters)
print("Number of noise points:", n_noise)
```
